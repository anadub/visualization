For this project, I used real-world data from Statistics Canada and created distinct visualizations to explore and communicate trends in operational costs and sales within Ontario’s greenhouse sector, and also show the difference between average payroll in Ontario vs Canada. These visualizations are based on publicly available datasets from Ontario’s Open Data Catalogue: https://data.ontario.ca/dataset/d14a058c-78bc-473c-8abf-79023bf7916b/resource/f1aebf69-f32e-4ebb-b310-8f010dc84ed7/download/greenhouse1_en.xlsx Source: Statistics Canada, CANSIM tables: 32-10-0019-01; 32-10-0022-01; 32-10-0023-01; 32-10-0026-01; 32-10-0245-01. 
I used Python for the first visualization and Google Sheets for the second.



```python
# Python script to visualize greenhouse industry costs and return in Ontario (2014–2024)

import pandas as pd
import matplotlib.pyplot as plt

fig, (costs, summary) = plt.subplots(ncols = 2,
                                   nrows = 1,
                                   figsize=(20, 8))


# Data
df_costs = pd.DataFrame({
    'Year': list(range(2014, 2025)),
    'Payroll': [27.04, 28.63, 27.55, 28.30, 27.97, 29.70, 29.51, 31.01, 31.90, 35.44, 33.10],
    'Purchase': [21.27, 21.63, 21.72, 22.12, 21.48, 21.18, 20.87, 20.64, 22.46, 24.55, 24.92],
    'Fuel': [11.60, 10.98, 9.13, 9.51, 9.47, 8.92, 8.04, 8.48, 8.74, 9.41, 8.92],
    'Electricity': [2.97, 3.12, 3.30, 3.08, 2.97, 2.75, 2.69, 3.38, 3.43, 4.45, 4.54]
})

# Plot setup

for col, color, shape in zip(['Payroll', 'Purchase', 'Fuel', 'Electricity'],
                      ['#0072B2', '#E69F00', '#009E73', '#CC79A7'],
                      ['*', 's', 'D', '^']):  # Circle, square, diamond, triangle_up
    costs.plot(df_costs['Year'], df_costs[col], label=col, color=color, markersize = 12, linewidth=2, marker=shape)


costs.set_title("Yes, most main operational costs are rising, but...", fontsize=16, weight='bold')
costs.set_xlabel("Year")
costs.set_ylabel("Cost per 1,000 sqm ($'000)")
costs.grid(True, linestyle='--', alpha=0.6)
costs.legend(title="Cost Categories")
costs.set_xticks(df_costs['Year'])


# Data
df_summary = pd.DataFrame({
    'Year': list(range(2014, 2025)),
    'Total main operational costs': [62.87, 64.35, 61.70, 63.01, 61.89, 62.55, 61.11, 63.50, 66.54, 73.85, 71.48],
    'Total Sales': [121.15,	122.89,	116.73,	117.31,	117.30,	124.59,	122.96,	124.69,	133.68,	151.11,	151.49]
})

# Plot setup
for col, color, marker in zip(['Total main operational costs', 'Total Sales'],
                      ['#D55E00', '#56B4E9'],
                      ['X', 'o']):  # X and star markers
    summary.plot(df_summary['Year'], df_summary[col], label=col, color=color, markersize = 12, linewidth=2.8, marker=marker)


summary.set_title("sales are rising more! ", fontsize=16, weight='bold')
summary.set_xlabel("Year")
summary.set_ylabel("Amount per 1,000 sqm ($'000)")
summary.set_ylim(bottom=0)  # This forces the Y-axis to start at 0
summary.grid(True, linestyle='--', alpha=0.6)
summary.legend(title="Categories")
summary.set_xticks(df_summary['Year'])

plt.subplots_adjust(wspace=4.7)  # Increase horizontal spacing between subplots

# Set font properties 
import matplotlib as mpl

mpl.rcParams.update({
    'font.size': 14,
    'font.family': 'Courier New',  # it is monospaced and accessible
    'axes.titlesize': 14,
    'axes.labelsize': 14,
    'xtick.labelsize': 14,
    'ytick.labelsize': 14,
    'legend.fontsize': 14,
    'legend.title_fontsize': 14
})

plt.tight_layout()
fig.text(0.5, -0.02, "Source: Statistics Canada, CANSIM tables: 32-10-0019-01; 32-10-0022-01; 32-10-0023-01; 32-10-0026-01; 32-10-0245-01",
         ha='center', fontsize=12, fontfamily='Courier New')

plt.show()
```

Key Messages and Insights

The first Python visualization is a side-by-side plot of major operational costs in Ontario’s greenhouse industry from 2014 to 2024. It highlights the evolution of payroll, purchase inputs, fuel, and electricity, showing an upward trend in all but fuel. The second subplot summarizes total sales versus total main operational costs. The main insight is that since 2022, and for the first time in many years (except for 2020), total sales have more than doubled total main operational costs, despite of its upward tendency since 2020. This suggests improved returns on operational investments.

It is important to note that the source doesn't mention other operational costs. Therefore, I assumed the selected categories represent the main sources of operational expenses. Also, due to the absence of data on machinery and installation acquisition or maintenance - for example - we cannot make conclusions about net gains.

The Google Sheets visualization compares the average payroll per employee in Ontario vs Canada between 2014 and 2023 (2024 data was not available). Ontario's payroll costs are rising and consistently outpacing the national average by approximately 13% to 16%. 

Design Considerations

For the Python visualizations, I considered:

Telling a short story addressing the main concern of the stakeholders.

Using color-blind–friendly palettes (e.g., #0072B2, #E69F00) .

Including distinct marker shapes to aid users with low vision.

Adding grid lines and large fonts to enhance clarity.

Choosing a monospaced, accessible font (Courier New) for better legibility.

Applying whitespace thoughtfully to avoid visual clutter.


For the Google Sheets visualization, I used:

Side-by-side bar charts to display provincial vs national trends.

Percentage difference overlays to highlight the scale of divergence.

A consistent color scheme and large, readable labels and fonts for accesibility.

More granular sales data, type of greenhouse, and total number of employees were excluded to reduce noise. Total area is implicitily used as the metrics are per '000 sqm.

In the Google Sheets visualization, I included provincial and national payroll data to offer comparative insight. The average payroll per employee was calculated.
