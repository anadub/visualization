```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
# Set a random seed for reproducibility
np.random.seed(613)
# Generate some random data for the scatter plot
# This code generates two sets of random integers for the y-axis values 
# and a range of integers for the x-axis values.
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
 
```

```python
# Plot both sets of data on the same plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2)
fig.show()
 
```

```python
# Add labels and a legend to the plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right') # Add a legend to the plot
fig.show()
 
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right',
     frameon = True, #add frame around the legend
     fontsize = 12, #change font size
     ncol = 2, #specify number of columns
     shadow = True ) 
fig.show()
 
```

```python
# We can also place the legend outside the plot area
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left', # where we are anchoring the legend relative to the plot
     bbox_to_anchor =(1, 1)) 
fig.show()
 
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower left',
     bbox_to_anchor =(1, 1))
fig.show()
 
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text(10,95,"This value is important!")
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text(10, 95, "This value is important!",
   ha='center',    # alignment of text
   color = 'red',  # modify font colour
   size = 20)      # modify font size
fig.show()
```

```python
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

ax.text(1, 5, ". Data:(1, 5)",           transform=ax.transData    ) # data coordinates
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)",   transform=ax.transAxes    ) # axes coordinates
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure ) # figure coordinates
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94), # xy coordinates of the annotation and text
     arrowprops=dict(facecolor='purple')) # Add an annotation with an arrow
fig.show()
 
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!',
   xy=(10, 95), xytext=(20, 94),
   arrowprops = dict(arrowstyle = "wedge",
   color = 'hotpink'))
fig.show()
 
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator()) # Remove y-axis ticks
ax.xaxis.set_major_formatter(plt.NullFormatter()) # Remove x-axis ticks
 
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MaxNLocator(3)) # Set the x-axis to have a maximum of 3 ticks
 
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MultipleLocator(5)) 
# Set the x-axis to have ticks at every 5 units
 
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

plt.xticks(rotation=45, ha='right', fontsize=12, font='serif') 
# Rotate x-axis labels for better readability
fig.show()
```

```python
font1 = {'family':'serif','color':'indigo'}
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xlabel('Shiny New X Axis!', fontsize = 14, fontdict = font1)
plt.show()
 
```
