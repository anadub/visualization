```python
font1 = {'family': 'sans-serif', 'color':'blue', 'size': 20}
font2 = {'family': 'monospace', 'color':'green', 'size': 14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict = font1, loc = 'left') # loc = 'left' aligns the title to the left
ax.set_ylabel('Total growth', fontdict = font2  )
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd
 
# y-axis in bold
rc('font', weight='bold')
 
# Values of each group
bars1 = [12, 28, 1, 8, 22]
bars2 = [28, 7, 16, 14, 10]
bars3 = [25, 3, 23, 25, 17]
 
# Heights of bars1 + bars2
bars = np.add(bars1, bars2).tolist()
 
# The position of the bars on the x-axis
r = [0,1,2,3,4]
 
# Names of group and bar width
names = ['A','B','C','D','E']
barWidth = 1
 
# Create brown bars
plt.bar(r, bars1, color="#eb1220", edgecolor='white', width=barWidth)
# Create green bars (middle), on top of the first ones
plt.bar(r, bars2, bottom=bars1, color="#3d2d7f", edgecolor='white', width=barWidth)
# Create green bars (top)
plt.bar(r, bars3, bottom=bars, color="#14be2d", edgecolor='white', width=barWidth)
 
# Custom X axis
plt.xticks(r, names, fontweight='bold')
plt.xlabel("group")
 
# Show graphic
plt.show()

```
